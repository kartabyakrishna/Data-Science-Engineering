```python
import os

def print_directory_structure(base_dir, indent=''):
    # List all files and subdirectories in the current directory
    items = os.listdir(base_dir)

    for item in items:
        item_path = os.path.join(base_dir, item)

        # Check if it's a directory
        if os.path.isdir(item_path):
            print(indent + f"ğŸ“ {item}/")
            print_directory_structure(item_path, indent + "  ")  # Recursive call

        # Check if it's a file
        elif os.path.isfile(item_path):
            print(indent + f"ğŸ“„ {item}")

# Get the current working directory
base_directory = os.getcwd()

print(f"ğŸ“ {base_directory}/")
print_directory_structure(base_directory, "  ")
```

```python
import os

def print_directory_structure(base_dir, indent=''):
    # List all files and subdirectories in the current directory, excluding hidden files
    items = [item for item in os.listdir(base_dir) if not item.startswith('.')]

    for item in items:
        item_path = os.path.join(base_dir, item)

        # Check if it's a directory
        if os.path.isdir(item_path):
            print(indent + f"ğŸ“ {item}/")
            print_directory_structure(item_path, indent + "  ")  # Recursive call

        # Check if it's a file
        elif os.path.isfile(item_path):
            print(indent + f"ğŸ“„ {item}")

# Get the current working directory
base_directory = os.getcwd()

print(f"ğŸ“ {base_directory}/")
print_directory_structure(base_directory, "  ")
```
