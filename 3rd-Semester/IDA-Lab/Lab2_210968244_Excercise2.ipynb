```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df.index += 1
df
```

```python
#Q1. Find the car with the best mpg.
```

```python
Best_mpg=df[df['mpg'] == df['mpg'].max()] 
print("Car with best mpg is \n" + str(Best_mpg['model']))
```

```python
#Q2. Find the car with the worst mpg.
```

```python
Worst_mpg=df[df['mpg'] == df['mpg'].min()] 
print("Car(s) with worst mpg is/are \n" + str(Worst_mpg['model']))
```

```python
#Q3. Find the car with the best horsepower.
```

```python
Best_HP=df[df['hp'] == df['hp'].max()] 
print("Car with best HP is \n" + str(Best_HP['model']))
```

```python
#Q4. Find 5 number summary of displacement.
```

```python
print("Here's the summary of displacement data : \n")
print(round(df['disp'].describe(),3))
```

```python
print("Min : ",np.min(df['disp']))
print("Max : ",np.max(df['disp']))
print("Std : ",round(np.std(df['disp']),3))
print("Mean : ",round(np.mean(df['disp']),3))
print("Median : ",np.median(df['disp']))
```

```python
#Q5. Find median horse power.
```

```python
print("Median Horse Power is \n" , (df.hp).median())
```

```python
#Q6.  What is average mpg for manual vs. automatic cars ?
```

```python
AutomaticCars = df[df['am'] == 1] 
print( "Average of MPG of Automatic cars is",  round(AutomaticCars['mpg'].mean(),4),".")
ManualCars = df[df['am'] == 0] 
print( "Average of MPG of Manual cars is",  round(ManualCars['mpg'].mean(),4),".")
```

```python
#Q7. Draw a histogram of miles per gallon .
```

```python
df['mpg'].hist(grid=False, bins=10)
plt.xlabel('MPG')
plt.ylabel('# of Cars')
plt.title('MPG Distribution')
```

```python
#Q8. Boxplot of mpg for each cylinder type
```

```python
sb.boxplot(x="cyl", y="mpg", data=df, palette="Set1", linewidth=2)
```
