```python
def cont_to_exp(x):
    if type(x) is str:
        x = x.replace('\\','')
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

### Q5. Add the Polarity, length of the review, the word count and average word length of each review.

```python
from textblob import TextBlob
```

```python
# polarity column
df['polarity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)

# review length column
df['review_len'] = df['review'].apply(lambda x:len(x))

# word count column
df['word_count'] = df['review'].apply(lambda x: len(x.split()))
```

```python
def avg_word_len(x):
    words = x.split()
    wordlen = 0
    for word in words:
        wordlen = wordlen + len(word)
        
    return wordlen/len(words) 
```

```python
df['avg_word_len'] = df['review'].apply(lambda x: avg_word_len(x))
```

### Q11. Visualize the distribution of Top 25 Unigram, Bigram and Trigram.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def top_words(x, n=25):
    vec = CountVectorizer(stop_words='english').fit(x)
    bag_of_words = vec.transform(x)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
words = top_words(df['review'], 25)
words
```

```python
dfu = pd.DataFrame(words, columns = ['UnigramText' , 'Count'])
dfu
```

```python

plot=dfu.plot.bar(x='UnigramText',y='Count',figsize=(17,7),width =0.75)
plt.legend(fontsize=15,loc='upper right')
plot.bar_label(plot.containers[0])
plt.xlabel('unigrams',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.xticks(fontsize=12)
plt.title('Unigrams Count',fontsize=15)
plt.show()
```

```python
def get_top_n_bigram(x, n=25):
    vec = CountVectorizer(ngram_range=(2, 2), stop_words='english').fit(x)
    bag_of_words = vec.transform(x)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
common_words = get_top_n_bigram(df['review'], 25)
common_words
```

```python
dfb = pd.DataFrame(common_words, columns = ['BigramText' , 'Count'])
dfb
```

```python
plot=dfb.plot.bar(x='BigramText',y='Count',figsize=(17,7),width =0.75)
plt.legend(fontsize=15,loc='upper right')
plot.bar_label(plot.containers[0])
plt.xlabel('bigrams',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.xticks(fontsize=12)
plt.title('bigrams Count',fontsize=15)
plt.show()
```

```python
def get_top_n_trigram(x, n=25):
    vec = CountVectorizer(ngram_range=(3, 3), stop_words="english").fit(x)
    bag_of_words = vec.transform(x)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
common_words = get_top_n_trigram(df['review'], 25)
common_words
```

```python
dft = pd.DataFrame(common_words, columns = ['TrigramText' , 'Count'])
dft
```

```python
plot=dft.plot.bar(x='TrigramText',y='Count',figsize=(17,7),width =0.75)
plt.legend(fontsize=15,loc='upper right')
plot.bar_label(plot.containers[0])
plt.xlabel('Trigrams',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.xticks(fontsize=12)
plt.title('Trigrams Count',fontsize=15)
plt.show()
```
